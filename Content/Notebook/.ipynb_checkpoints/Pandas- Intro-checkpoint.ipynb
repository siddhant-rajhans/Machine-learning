# Modiying Price by removing `$ and ,`

```python
# Remove dollar sign and comma
sales["Price"] = sales["Price"].str.replace('$', '').str.replace(',', '')

# Convert the cleaned string to float (handles decimals)
sales["Price"] = sales["Price"].astype(float)

# Now you can work with the "Price" column as a numeric column
sales
```

## Dealing with missisng value dataset

```python
msales= pd.read_csv("./data/car-sales-missing-data.xls")
```

```python
fmsales= pd.read_csv("./data/car-sales-missing-data.xls")
```

```python
msales
```

```python
#msales["Odometer"] = msales["Odometer"].fillna(msales["Odometer"].mean(), inplace= True)
fmsales["Odometer"] = fmsales["Odometer"].fillna(fmsales["Odometer"].mean())
```

```python
fmsales
```

## Fill NaN values in "Odometer" with the previous valid value (ffill)
    -msales["Odometer_FFill"] = msales["Odometer"].fillna(method='ffill')
## Fill NaN values in "Odometer" with the next valid value (bfill)
    -msales["Odometer_BFill"] = msales["Odometer"].fillna(method='bfill')



# To drop the rows containing missing data

```python
#dropped missing sales - dmsales
#dmsales = fmsales.dropna()
fmsales.dropna(inplace = True) 
# To reverse this- must reimport dataset
```

```python
fmsales
```
