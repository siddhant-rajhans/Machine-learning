## Matplotlib eg workflow:

```python
# Import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4] 
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,5)) # width, height

# 3. Plot the data
ax.plot(x,y)
 
# 4. Customize plot 
ax.set(title= "My new plot", 
      xlabel = "X-axis",
      ylabel = "y-axis")

# 5. Save & show (saving the figure)
fig.savefig("images/myplot.png")
```

## Tow options for subplot


```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows= 2, ncols=2,figsize=(10,5))


# Plot tot each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter.keys(), butter.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot optrion 2
fig, ax = plt.subplots(nrows =2, ncols= 2, figsize =(10,5))

# Plot to each figgerent  
# Index the ax variable to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(butter.keys(), butter.values());
ax[1, 1].hist(np.random.randn(1000));
```

# 4. Plotting more advanced plots from a pandas DataFrame
It's possible to achieve far more complicated and detailed plots from a pandas DataFrame.

Let's practice using the heart_disease DataFrame.

* For quick plot use pyplot
* For advance plot use OO method

```python
# heartd["50+"] = heartd["age"]
# heartd
```

```python
type(heartd.age)
heartd
```

```python
old = heartd[heartd['age'] >50]
old
```

```python
# Pyplot method
old.plot(kind='scatter', x='age', y= 'chol', c= 'target');
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(20,12))
old.plot(kind='scatter', x= 'age', y = 'chol', c='target', ax=ax)
ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(x= old["age"], y= old['chol'],c = old['target'] )

#  Customize 
ax.set(title="Heart disease & Cholesterol",
      xlabel= 'Age',
      ylabel= 'Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add horizontal line
ax.axhline(old["chol"].mean(),
          linestyle='--');
```

```python
old.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(nrows=2,
                             ncols=1,
                             figsize= (10,10))
# Add data to ax0
scatter = ax0.scatter(x=old['age'], y=old['chol'], c=old['thalach'])
ax0.set(title= 'Heart Disease',
       xlabel= "Age",
       ylabel= "Cholesterol")

# Add a legend top ax0
ax0.legend(*scatter.legend_elements(), title= "Target");

# Add data to ax1
scatter = ax1.scatter(x=old['age'], y=old['thalach'], c=old['target'])
ax1.set(title= 'Heart Disease',
       xlabel= "Age",
       ylabel= "Max Heart rate")

# Add a legend top ax0
ax1.legend(*scatter.legend_elements(), title= "Target");
```
